```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#importando base de dados

file_csv = 'b1abb33c4f.csv'

#carregando arquivo em um dataframe
data = pd.read_csv(file_csv)

# Exibir as primeiras linhas do DataFrame
print(data.head())
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
data[['IDADEMAE']].boxplot()
```

```python
# Criar o boxplot para a idade das mães
sns.boxplot(x=data['IDADEMAE']) #cria o gráfico diretamente 
plt.title('Boxplot da Idade das Mães') #título do gráfico
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Criar o boxplot para o peso do bebê
print(data.columns)
```

```python
# Criar o boxplot para o peso bebê
sns.boxplot(x=data['PESO']) #cria o gráfico diretamente 
plt.title('Boxplot Peso do Bebê') #título do gráfico
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python

# Criar o histograma para a idade das mães

plt.figure(figsize=(8, 6))
plt.hist(data['IDADEMAE'].dropna(), bins=10, color='skyblue', edgecolor='black') #Remove valores ausentes
plt.title('Histograma da Idade das Mães', fontsize=16)
plt.xlabel('Idade das Mães', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(8, 6))
plt.hist(data['IDADEMAE'].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma Peso do Bebê', fontsize=16)
plt.xlabel('Peso do Bebê', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.show()
```
